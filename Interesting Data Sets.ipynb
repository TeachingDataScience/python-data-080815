```python
import pandas as pd
```

```python
# LAPD Crime and Collission Data
df = pd.read_csv("https://data.lacity.org/api/views/eta5-h8qx/rows.csv?accessType=DOWNLOAD")
```

```python
df.dtypes
```

```python
df[df["Crm Cd Desc"] == "ROBBERY"]
```

```python
CrimeCounts = {}

for i in df["Crm Cd Desc"].unique():
    CrimeCounts[i] = len(df[df["Crm Cd Desc"] == i])
```

```python
CrimeCounts
```

```python
CC = pd.Series(CrimeCounts)
```

```python
%matplotlib inline
```
